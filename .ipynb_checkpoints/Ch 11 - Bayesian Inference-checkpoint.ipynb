```python
%matplotlib inline
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.stats import beta
```

### 2

```python
mu = 5
n = 100
X = np.random.normal(mu, 1, size=(n, ))
```

$f(\theta|X) \propto p(\theta) \mathcal{L}(X) \propto \text{exp}\left(-\frac{1}{2}\sum_{i=1}^n(x_i-\mu)^2\right) \propto \text{exp}\left (\frac{n}{2}(\mu -\bar X)^2 \right )$. So, we have posterior
$$
f(\theta|X) \sim \mathcal{N}(\bar X, \frac{1}{n})
$$

where $\bar X = \frac{1}{n} \sum_{i=1}^n X_i$

```python
f = lambda u: np.sqrt(n/2*math.pi) * np.exp(-n/2*(u - X.mean())**2)

lu = np.linspace(X.mean()-0.5, X.mean()+0.5, 300)
plt.plot(lu, f(lu), 'b')
plt.title("Posterior Distribution")
plt.xlabel("u")
```

```python
# sample 1000 points from posterior

Usample = np.random.normal(X.mean(), 1/np.sqrt(n), 1000)

plt.hist(Usample, bins=30, density=True)
plt.title("Sample Posterior Distribution")
plt.xlim(4.6, 5.6)
plt.plot(lu, f(lu))
```

$$
f(\mu) = \sqrt {\frac{n}{2\pi}} e^{-\frac{n}{2}(\mu-\bar X)^2}
$$

Then $\theta = e^\mu$. We have
$$
\begin{align}
F(\theta < x) &= F(e^\mu < x) \\
&= F(\mu < \log x) \\
&= \Phi(\log x)
\end{align}
$$
Therefore, 
$$
\begin{align}
P(\theta = x) &= F'\left(\theta < x\right) \\
&= \frac{d \Phi(\log x) }{d x} \\
&= \frac{1}{x}f(\log x) \\
&= \frac{1}{x} \sqrt {\frac{n}{2\pi}} e^{-\frac{n}{2}(\log x-\bar X)^2}
\end{align}
$$


```python
# theoretical 
theta = lambda x: 1/x * np.sqrt(n/2*math.pi) * np.exp(-n/2*(np.log(x) - X.mean())**2)

x_theta = np.linspace(np.exp(4.6), np.exp(5.6), 1000)
plt.plot(x_theta, theta(x_theta))
```

```python
# sample mu

U_sample = np.random.normal(X.mean(), 1/np.sqrt(n), 5000)
theta_sample = np.exp(U_sample)

plt.hist(theta_sample, bins=30, density=True)
plt.plot(x_theta, theta(x_theta))
```

```python
# 95% interval for mu
print("95 percent interval for mu: (%.4f, %.4f)"%(X.mean()-1.96/np.sqrt(n),X.mean()+1.96/np.sqrt(n)))
```

### 5

Data: $s=2, n=10$.

Therefore, the posterior 
$$
p(\theta|X) = \text{Beta}(\alpha + s, \beta + n - s )
$$

So basically, as $\alpha$ and $\beta$ increase, there is less influence of data on the posterior.

```python
s = 2
n = 10


def beta_fun(a, b):
    x = np.linspace(beta.ppf(0.01, a+s, b+n-s), beta.ppf(0.99, a+s, b+n-s), 100)
    plt.plot(x, beta.pdf(x, a+s, b+n-s),'r-', lw=5, alpha=0.6, label='beta pdf')
    plt.title("alpha=%s, beta=%s"%(a, b))
```

```python
beta_fun(1/2, 1/2)
```

```python
beta_fun(1, 1)
```

```python
beta_fun(10, 10)
```

```python
beta_fun(100, 100)
```
